```python
#Import library yang akan gigunakan
import pandas as pd #Library panda untuk membaca file csv
from sklearn.model_selection import train_test_split #Train/test split adalah salah satu metode yang dapat digunakan untuk mengevaluasi performa model machine learning. Metode evaluasi model ini membagi dataset menjadi dua bagian yakni bagian yang digunakan untuk training data dan untuk testing data dengan proporsi tertentu.
from sklearn.preprocessing import StandardScaler #Library untuk melakukan tugas Standardisasi . Biasanya dataset berisi variabel yang berbeda dalam skala misal umur ada yang 48-60
scaler=StandardScaler() #Inisiasi fungsi scaler
```

```python
#load data
url = 'https://raw.githubusercontent.com/Nabilahfzh17/Pembelajaran-Mesin/main/diabetes_data_upload.csv'
df = pd.read_csv(url, sep=',')

df.head()
```

```python
df.describe(include='all') #Menunjukkan rangkuman statistik seperti rata-rata, median, dll pada kolom.
```

```python
#mendefinisikan fungsi untuk preprosessing data
def preprocessing(data):
    df= data.copy() #menyalin dataFrame yang disimpan di parameter data
    
    # Mengubah isi kolom gender yang semula female dan male menjadi 0 dan 1
    df['Gender'] = df ['Gender'].replace({'Female':0,'Male':1 })
    
    #Mengubah isi kolom selain Age, Gender, Class yang berisi 'No' dan 'Yes' menjadi 0 dan 1
    for column in df.columns.drop(['Age','Gender','class']):
        df[column]= df[column].replace({'No':0 , 'Yes': 1})
    
    #Melakukan Traning
    y=df["class"] #menyimpan kolom class ke variabel y
    X=df.drop("class", axis=1) #drop kolom class 
    
    #test_train_split membagi data menjadi dua bagian yaitu traning dan test
    X_train, X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,shuffle=True,random_state=1)
    
    #StandardScaler
    scaler.fit(X_train) #Hitung mean dan std yang akan digunakan untuk penskalaan nanti.
    X_train=pd.DataFrame(scaler.transform(X_train),index=X_train.index , columns=X_train.columns) #membuat dataframe menggunakan X_train
    X_test=pd.DataFrame(scaler.transform(X_test),index=X_test.index, columns=X_test.columns) #membuat dataframe menggunakan X_test
    
    return X_train,X_test,y_train,y_test #mereturn X_train,X_test, y_train, y_test ketika fungsi preprocessing di panggil
```

```python
#memanggil fungsi preprocessing yang sudah di buat tadi dengan argumen data (variabel yang menyimpan dataset .csv)
X_train,X_test,y_train,y_test= preprocessing(df)
```

```python
#menampilkan dataframe X_train
X_train
```

```python
#import library KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2) #settingan default knn
knn.fit(X_train, y_train) #setelah membuat data train dan test, maka masukkan data training pada fungsi klasifikasi KNN
```

```python
y_pred = knn.predict(X_test) #melakukan prediksi ke data test 
y_pred
```

```python
#menghitung matrix confusion dari test dengan prediksi klasifikasi knn
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

Dari model knn yang kita buat berdasarkan matrik confusion kita mendapat 
True Positif = 54, False Positif = 4, True Negatif = 87, False Negatif = 11
